```python
import re
import string
import docx2txt

frequency = {}
document_text = docx2txt.process("testdoc.docx")
match_pattern = re.findall(r'\b[a-zåøæäöü]{1,99}\b', document_text, re.I)

for word in match_pattern:
    count = frequency.get(word,0)
    frequency[word] = count + 1
     
frequency_list = sorted(frequency.keys())
 
print("Here are all words in the text, in alphabetical order, with their frequency-of-occurrence:")
for words in frequency_list:
    print(words.lower(),"-->", frequency[words],"times")
```
