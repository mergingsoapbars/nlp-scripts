```python
#this script uses an ordered dictionary of all words in a docx.-file and sorts them according to their frequencies.

import docx2txt
import collections
import nltk
from nltk.tokenize import RegexpTokenizer

tokenizer = RegexpTokenizer('[A-Za-z]+')

#replace the document name with your own -- make sure python can locate it
document_text = docx2txt.process("testdoc.docx")

tokens = tokenizer.tokenize(document_text)
tokens_lower = [word.lower() for word in tokens]

freq = nltk.FreqDist(tokens_lower)

freq_dict = dict(freq)
sorted_dic = collections.OrderedDict(sorted(freq_dict.items()))
freq_t10 = freq.most_common()
freq_t10_dic = dict(freq_t10)

print("The following is a list of all the words in the text, sorted alphabetically, and their respective frequencies:")
for key in freq_t10_dic:
    print(key,"-->", freq_t10_dic[key],"times")
```
