```python
#this script provides an ordered dictionary of all words in a docx.-file

import docx2txt
import collections
import nltk
from nltk.tokenize import RegexpTokenizer

tokenizer = RegexpTokenizer('[A-Za-z]+')

#replace the document name with your own -- make sure python can locate it
document_text = docx2txt.process("testdoc.docx")

tokens = tokenizer.tokenize(document_text)
tokens_lower = [word.lower() for word in tokens]

freq = nltk.FreqDist(tokens_lower)

thing = dict(freq)
sorted_dic = collections.OrderedDict(sorted(thing.items()))

print("The following is a list of all the words in the text, sorted alphabetically, and their respective frequencies:")
for key in sorted_dic:
    print(key, "-->", sorted_dic[key], "times")
```
