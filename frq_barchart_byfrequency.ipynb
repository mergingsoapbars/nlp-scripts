```python
import docx2txt
import re
from matplotlib import pyplot as plt
```

```python
frequency = {}
document_text = docx2txt.process("testdoc.docx")
match_pattern = re.findall(r'\b[a-zåøæäöü]{1,99}\b', document_text, re.I)
```

```python
for word in match_pattern:
    count = frequency.get(word,0)
    frequency[word] = count + 1
```

```python
frequency_sorted = sorted(frequency.items(), key=lambda x: x[1], reverse = True)
```

```python
frequency_sorted_top = dict(frequency_sorted[:50])
```

```python
word_list = frequency_sorted_top.keys()
frequency_list = frequency_sorted_top.values()
```

```python
plt.figure(figsize=(20.0,20.0))
plt.xticks(
    rotation=90, 
    fontsize=12  
)
plt.bar(word_list, frequency_list)
plt.show()
```
