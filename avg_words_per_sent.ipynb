```python
import nltk
from nltk.tokenize import sent_tokenize
import docx2txt
import re
```

```python
#replace the document name with your own -- make sure python can locate it
doc_text = docx2txt.process("testdoc.docx")
```

```python
sentences = sent_tokenize(doc_text)
```

```python
llist = []
```

```python
for sentence in sentences:
    res = len(sentence.split())
    llist.append(res)
```

```python
lsum = sum(llist)
llength = len(llist)
```

```python
print("There are", listsum, "words and", llength, "sentences in the document, which puts the average number of words per sentence at",(lsum / llength))
```
